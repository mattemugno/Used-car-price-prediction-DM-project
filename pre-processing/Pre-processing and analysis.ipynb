```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
df = pd.read_csv("C://Users//matte//OneDrive//Desktop//data mining//datasetAutoScout//scrapingAutoScout.csv", encoding='unicode_escape', delimiter =",")
df.info()
```

```python
# BINARIZATION COMFORT AND CONVENIENCE
list = []
for index,value in df['Comfort & Convenience'].items():
    comf_conv = str(value).split("; ")[:-1]
    for i in comf_conv:
        if i not in list:
            list.append(i)

contatore = [0 * i for i in range(len(list))]
for index, value in df['Comfort & Convenience'].items():
    comf_conv = str(value).split("; ")[:-1]
    for i in comf_conv:
        contatore[list.index(i)] += 1

sorted_list = [b[0] for b in sorted(enumerate(contatore),key=lambda i:i[1])]
sorted_list.reverse()

i = 0
top_comf_conv = []
for value in sorted_list:
    if i >= 5:
        break
    top_comf_conv.append(list[value])
    df[list[value]] = 0
    i += 1

for index, value in df['Comfort & Convenience'].items():
    comf_conv = str(value).split("; ")[:-1]
    for i in comf_conv:
        if i in top_comf_conv:
            df[i].at[index] = 1
```

```python
# BINARIZATION ENTERTAINMENT AND MEDIA
list = []
for index,value in df['Entertainment & Media'].items():
    ent_media = str(value).split("; ")[:-1]
    for i in ent_media:
        if i not in list:
            list.append(i)

contatore = [0 * i for i in range(len(list))]
for index, value in df['Entertainment & Media'].items():
    ent_media = str(value).split("; ")[:-1]
    for i in ent_media:
        contatore[list.index(i)] += 1

sorted_list = [b[0] for b in sorted(enumerate(contatore),key=lambda i:i[1])]
sorted_list.reverse()

i = 0
top_ent_media = []
for value in sorted_list:
    if i >= 5:
        break
    top_ent_media.append(list[value])
    df[list[value]] = 0
    i += 1

for index, value in df['Entertainment & Media'].items():
    ent_media = str(value).split("; ")[:-1]
    for i in ent_media:
        if i in top_ent_media:
            df[i].at[index] = 1
```

```python
# BINARIZATION EXTRAS
list = []
for index,value in df['Extras'].items():
    extras = str(value).split("; ")[:-1]
    for i in extras:
        if i not in list:
            list.append(i)

contatore = [0 * i for i in range(len(list))]
for index, value in df['Extras'].items():
    extras = str(value).split("; ")[:-1]
    for i in extras:
        contatore[list.index(i)] += 1

sorted_list = [b[0] for b in sorted(enumerate(contatore),key=lambda i:i[1])]
sorted_list.reverse()

i = 0
top_extras = []
for value in sorted_list:
    if i >= 5:
        break
    top_extras.append(list[value])
    df[list[value]] = 0
    i += 1

for index, value in df['Extras'].items():
    extras = str(value).split("; ")[:-1]
    for i in extras:
        if i in top_extras:
            df[i].at[index] = 1
```

```python
#BINARIZATION SAFETY AND SECURITY
list = []
for index,value in df['Safety & Security'].items():
    safe_sec = str(value).split("; ")[:-1]
    for i in safe_sec:
        if i not in list:
            list.append(i)

contatore = [0 * i for i in range(len(list))]
for index, value in df['Safety & Security'].items():
    safe_sec = str(value).split("; ")[:-1]
    for i in safe_sec:
        contatore[list.index(i)] += 1

sorted_list = [b[0] for b in sorted(enumerate(contatore),key=lambda i:i[1])]
sorted_list.reverse()

i = 0
top_safe_sec = []
for value in sorted_list:
    if i >= 5:
        break
    top_safe_sec.append(list[value])
    df[list[value]] = 0
    i += 1

for index, value in df['Safety & Security'].items():
    safe_sec = str(value).split("; ")[:-1]
    for i in safe_sec:
        if i in top_safe_sec:
            df[i].at[index] = 1
```

```python
# DROP SELLER AND FULL SERVICE HISTORY
df.drop(df.columns[28], axis = 1, inplace = True)
df.drop(df.columns[16], axis = 1, inplace = True)
```

```python
# MERGE DELLE 2 COLONNE DEL FUEL PER OTTENERE SU UNA SOLA COLONNA ANCHE LE MACCHINE ELETTRICHE E IBRIDE
# ALLA FINE SI PUO ELIMINARE OTHER FUEL TYPES
for index, value in df['Other fuel types'].items():
    if(pd.isna(value) is False) and (pd.isna(df['Fuel type'].at[index]) is False):
        df['Fuel type'].at[index] = 'Hybrid'
    if(pd.isna(value) is False) and (pd.isna(df['Fuel type'].at[index]) is True):
        df['Fuel type'].at[index] = 'Electric'
        df['CO2 Emissions'].at[index] = 0
df.drop(df.columns[21], axis = 1, inplace = True)
```

```python
# ELIMINAZIONE DI VARI CARATTERI
df["CO2 Emissions"] = df["CO2 Emissions"].str.replace("g/km","")
df["CO2 Emissions"] = df["CO2 Emissions"].str.replace(r'[(comb.)]',"")
df["CO2 Emissions"] = df["CO2 Emissions"].fillna(0).astype(float)

df["Empty weight"] = df["Empty weight"].str.replace("kg","")
df["Empty weight"] = df["Empty weight"].fillna(0).astype(int)

df["Engine size"] = df["Engine size"].str.replace("cc","")
df["Engine size"] = df["Engine size"].fillna(0).astype(int)

df["Mileage"] = df["Mileage"].str.replace("km","")
df["Mileage"] = df["Mileage"].fillna(0).astype(int)

df["Warranty"] = df["Warranty"].str.replace("months","")

df["Price"] = df["Price"].fillna(0).astype(int)
df["Previous owner"] = df["Previous owner"].fillna(0).astype(int)
df["Seats"] = df["Seats"].fillna(0).astype(int)
df["Doors"] = df["Doors"].fillna(0).astype(int)
df["Gears"] = df["Gears"].fillna(0).astype(int)
df["Cylinders"] = df["Cylinders"].fillna(0).astype(int)
df["CO2 Emissions"] = df["CO2 Emissions"].fillna(0).astype(int)

df["Fuel consumption"] = df["Fuel consumption"].str.replace("l/100 ","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace("kg/100","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace("(","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace("country","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace("city","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace("comb.","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace(")","")
df["Fuel consumption"] = df["Fuel consumption"].str.replace(" km",";")
df["Fuel consumption"] = df["Fuel consumption"].fillna(0)

df["Power"] = df["Power"].str.replace("hp","")
df["Power"] = df["Power"].str.replace("kW",";")
df["Power"] = df["Power"].str.replace("(","")
df["Power"] = df["Power"].str.replace(")","")
df["Power"] = df["Power"].fillna(0)
```

```python
# RIMPIAZZO I CONSUMI CON UN SOLO VALORE CHE RAPPRESENTA LA MEDIA DEI TRE
from statistics import mean
for index,value in df['Fuel consumption'].items():
    #print("Value-->" + str(value))
    if value != 0:
        nums = value.split(sep=';')
        my_list = []
        for i in range(0,len(nums)-1):
            my_list += [float(nums[i])]
        avg = round(mean(my_list), 1)
        df['Fuel consumption'].at[index] = avg
```

```python
# PARSING WARRANTY VALUES
for index,value in df['Warranty'].items():
    if value == "Yes":
        df['Warranty'].at[index] = 12
    if value == "No":
        df['Warranty'].at[index] = 0   
df["Warranty"] = df["Warranty"].fillna(0).astype(int)
```

```python
# COUNT OF NOMINAL ATTRIBUTE OF COMFORTS AND FEATURES
for index,value in df['Comfort & Convenience'].items():
    arr = str(value).split(sep=';')
    if pd.isna(arr[0]) is False:
        count = 0
        for elem in arr:
            count += 1
        df['Comfort & Convenience'].at[index] = count - 1
df["Comfort & Convenience"] = df["Comfort & Convenience"].fillna(0).astype(int)

for index,value in df['Entertainment & Media'].items():
    arr = str(value).split(sep=';')
    if pd.isna(arr[0]) is False:
        count = 0
        for elem in arr:
            count += 1
        df['Entertainment & Media'].at[index] = count - 1
df["Entertainment & Media"] = df["Entertainment & Media"].fillna(0).astype(int)

for index,value in df['Extras'].items():
    arr = str(value).split(sep=';')
    if pd.isna(arr[0]) is False:
        count = 0
        for elem in arr:
            count += 1
        df['Extras'].at[index] = count - 1
df["Extras"] = df["Extras"].fillna(0).astype(int)

for index,value in df['Safety & Security'].items():
    arr = str(value).split(sep=';')
    if pd.isna(arr[0]) is False:
        count = 0
        for elem in arr:
            count += 1
        df['Safety & Security'].at[index] = count - 1
df["Safety & Security"] = df["Safety & Security"].fillna(0).astype(int)
```

```python
# PRENDO SOLO IL VALORE DEI CAVALLI DALLA POTENZA
for index,value in df['Power'].items():
    if value != 0:
        hp = value.split(sep=';')[1]
        df['Power'].at[index] = hp
df["Power"] = df["Power"].astype(int)
```

```python
df = df[df['Model'].notna()]
df = df[df['Manufacturer'].notna()]
df = df[df['Fuel type'].notna()]
indexNames = df[ df['Mileage'] == 0 ].index
df.drop(indexNames , inplace=True)
indexNames = df[ df['Body type'] == 'Other' ].index
df.drop(indexNames , inplace=True)
```

```python
# CHANGE FOR SOME ATTRIBUTES
for index, value in df['Fuel type'].items():
    if "Diesel" in value:
        df['Fuel type'].at[index] = "Diesel"
    if ("Super" in value or "Gasoline" in value or "Other" in value or "Benz" in value or "Ethanol" in value):
        df['Fuel type'].at[index] = "Gasoline"
    if "LPG" in value or "Hydrogen" in value:
        df['Fuel type'].at[index] = "LPG"
    if "CNG" in value or "Domestic" in value:
        df['Fuel type'].at[index] = "Methane"
# doors
for index, value in df['Doors'].items():
    if value == 0 or value == 4:
        df['Doors'].at[index] = 5
    if value == 2:
        df['Doors'].at[index] = 3

# drivetrain
for index, value in df['Drivetrain'].items():
    if pd.isna(value):
        df['Drivetrain'].at[index] = "Rear"

# fuel consuption
for index, value in df['Fuel consumption'].items():
    if df['Fuel type'].at[index] == "Electric":
        df['Fuel consumption'].at[index] = 0

# gearbox 
for index, value in df['Gearbox'].items():
    if pd.isna(value):
        df['Gearbox'].at[index] = "Manual"

# gears
for index, value in df['Gears'].items():
    if pd.isna(value):
        if df['Gearbox'].at[index] == "Manual":
            df['Gears'].at[index] = 5
        if df['Gearbox'].at[index] == "Automatic":
            df['Gears'].at[index] = 7 



```

```python
# MISSING VALUES CO2 EMISSIONS
emission_df = df[((df['CO2 Emissions'] == 0) | (df['CO2 Emissions'].isna())) & (df['Fuel type'] != 'Electric')]    
for index,value in emission_df['Model'].items():
    
    model = emission_df['Model'].at[index]
    manufacturer = emission_df['Manufacturer'].at[index]
    body_type = emission_df['Body type'].at[index]
    fuel = emission_df['Fuel type'].at[index]
    engine = emission_df['Engine size'].at[index]
    
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['CO2 Emissions'] > 0) & (df['Fuel type'] == fuel) ]
    if result_df.empty:
        result_df = df[ (df['Body type'] == body_type) & (df['CO2 Emissions'] > 0)]
        df['CO2 Emissions'].at[index] = round(result_df['CO2 Emissions'].mean(),1)
    else:
        df['CO2 Emissions'].at[index] = round(result_df['CO2 Emissions'].mean(),1)
```

```python
# MISSING VALUES COLOUR
color_df = df[df['Colour'].isna()]
for index,value in color_df['Model'].items():
    model = color_df['Model'].at[index]
    manufacturer = color_df['Manufacturer'].at[index]
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (~df['Colour'].isna())]
    if result_df.empty:
        df['Colour'].at[index] = 'Black'
    else:  
        popular = result_df.groupby('Colour')['Colour'].count().sort_values().tail(1)
        colour = str(popular.keys()[0]).replace("'","")
        df['Colour'].at[index] = colour
```

```python
# emission class 
for index, value in df['Emission class'].items():
    if (df['Fuel type'].at[index] == "Electric") or (df['Fuel type'].at[index] == "Hybrid") :
        df['Emission class'].at[index] = "Euro 6"
```

```python
# MISSING VALUES CYLINDERS 

cylinders_df = df[df['Cylinders'] == 0]

for index,value in cylinders_df['Model'].items():
    model = cylinders_df['Model'].at[index]
    manufacturer = cylinders_df['Manufacturer'].at[index]
    power = cylinders_df['Power'].at[index]
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['Cylinders'] > 0)]
    if result_df.empty:
        result_df = df[(df['Power'] == power)]
        cylinders = round(result_df['Cylinders'].mean())
        df['Cylinders'].at[index] = cylinders
    else:  
        popular = result_df.groupby('Cylinders')['Cylinders'].count().sort_values().tail(1)
        cylinders = str(popular.keys()[0]).replace("'","")
        df['Cylinders'].at[index] = int(cylinders)   
df = df[df['Cylinders'] > 0]
```

```python
# PARSING FIRST REGISTRATION 
for index,value in df['First registration'].items():
    year = value.split("/")[1]
    df["First registration"].at[index] = year
    
for index,value in df['Emission class'].items():
    year = int(df['First registration'].at[index])
    if year < 1996:
        df['Emission class'].at[index] = "Euro 1"
    elif year >= 1996 and year < 2000:
        df['Emission class'].at[index] = "Euro 2"
    elif year >= 2000 and year < 2005:
        df['Emission class'].at[index] = "Euro 3"
    elif year >= 2005 and year < 2009:
        df['Emission class'].at[index] = "Euro 4"
    elif year >= 2009 and year < 2014:
        df['Emission class'].at[index] = "Euro 5"
    elif year >= 2014 :
        df['Emission class'].at[index] = "Euro 6"
```

```python
df = df[df['Body type'].notna()]
index = df[df['Country'].isna()].index
for values in index:
    df['Country'].at[values] = "DE"
```

```python
# MISSING VALUES EMPTY WEIGHT
weight_df = df[df['Empty weight'] == 0]

for index,value in weight_df['Empty weight'].items():
    
    model = weight_df['Model'].at[index]
    manufacturer = weight_df['Manufacturer'].at[index]
    body_type = weight_df['Body type'].at[index]
    
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['Empty weight'] > 0)]
    if result_df.empty:
        result_df = df[(df['Body type'] == body_type)]
        weight = round(result_df['Empty weight'].mean())
        df['Empty weight'].at[index] = weight
    else:  
        popular = result_df.groupby('Empty weight')['Empty weight'].count().sort_values().tail(1)
        weight = str(popular.keys()[0]).replace("'","")
        df['Empty weight'].at[index] = int(float(weight)) 
```

```python
# MISSING ENGINE SIZE
engine_df = df[df['Engine size'] == 0]

for index,value in engine_df['Engine size'].items():
    
    model = engine_df['Model'].at[index]
    manufacturer = engine_df['Manufacturer'].at[index]
    cylinders = engine_df['Cylinders'].at[index]
    
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['Engine size'] > 0)]
    if result_df.empty:
        result_df = df[(df['Cylinders'] == cylinders)]
        size = round(result_df['Engine size'].mean())
        df['Engine size'].at[index] = size
    else:  
        popular = result_df.groupby('Engine size')['Engine size'].count().sort_values().tail(1)
        size = str(popular.keys()[0]).replace("'","")
        df['Engine size'].at[index] = int(float(size))
```

```python
# MISSING VALUES POWER
power_df = df[df['Power'] == 0]
for index,value in power_df['Power'].items():
    
    model = power_df['Model'].at[index]
    manufacturer = power_df['Manufacturer'].at[index]
    cylinders = power_df['Cylinders'].at[index]
    
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['Power'] > 0)]
    if result_df.empty:
        result_df = df[(df['Cylinders'] == cylinders)]
        power = round(result_df['Power'].mean())
        df['Power'].at[index] = power
    else:  
        popular = result_df.groupby('Power')['Power'].count().sort_values().tail(1)
        power = str(popular.keys()[0]).replace("'","")
        df['Power'].at[index] = int(float(power))  
```

```python
# MISSING VALUES UPHOLSTERY COLOUR
upholstery_colour_df = df[df['Upholstery colour'].isna()]

for index, value in upholstery_colour_df['Upholstery colour'].items():
    df['Upholstery colour'].at[index] = 'Black'
```

```python
# MISSING VALUES UPHOLSTERY 
upholstery_df = df[df['Upholstery'].isna()]
for index,value in upholstery_df['Upholstery'].items():
    
    model = upholstery_df['Model'].at[index]
    manufacturer = upholstery_df['Manufacturer'].at[index]
    
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['Upholstery'].notna())]
    if result_df.empty:
        df['Upholstery'].at[index] = "Cloth"
    else:  
        popular = result_df.groupby('Upholstery')['Upholstery'].count().sort_values().tail(1)
        upholstery = str(popular.keys()[0]).replace("'","")
        df['Upholstery'].at[index] = upholstery
```

```python
# MISSING VALUES FUEL CONSUMPTION 
fuel_df = df[df['Fuel consumption'] == 0]
for index,value in fuel_df['Fuel consumption'].items():
    
    model = fuel_df['Model'].at[index]
    manufacturer = fuel_df['Manufacturer'].at[index]
    fuel_type = fuel_df['Fuel type'].at[index]
    body = fuel_df['Body type'].at[index]
    
    result_df = df[ (df['Manufacturer'] == manufacturer) & (df['Model'] == model) & (df['Fuel type'] == fuel_type) & (df['Fuel consumption'] > 0)]
    if result_df.empty:
        result_df = df[(df['Fuel consumption'] > 0) & (df['Body type'] == body) ]
        consumo = round(result_df['Fuel consumption'].mean(),1)
        df['Fuel consumption'].at[index] = float(consumo)
    else:  
        popular = result_df.groupby('Fuel consumption')['Fuel consumption'].count().sort_values().tail(1)
        fuel = str(popular.keys()[0]).replace("'","")
        df['Fuel consumption'].at[index] = float(fuel)
```

```python
# MISSING VALUES GEARS 
for index,value in df['Gears'].items():
    if value == 0 and df['Gearbox'].at[index] == "Automatic":
        df['Gears'].at[index] = int(7)
    elif value == 0 and df['Gearbox'].at[index] == "Manual":
        df['Gears'].at[index] = int(5)  
```

```python
# MISSING VALUES SEATS
for index,value in df['Seats'].items():
    if value == 0:
        df['Seats'].at[index] = int(5) 
```

```python
# MISSING VALUES UPHOLSTERY
for index,value in df['Upholstery'].items():
    if pd.isna(value):
        df['Upholstery'].at[index] = "Cloth"
```

```python
# MISSING VALUES UPHOLSTERY COLOUR
for index,value in df['Upholstery colour'].items():
    if pd.isna(value):
        df['Upholstery colour'].at[index] = "Black"
```

```python
# Trasformazione Gearbox in Nominal as Manual = 0 & Automatic = 1
for index,value in df['Gearbox'].items():
    if value == "Manual":
        df['Gearbox'].at[index] = int(1)
    elif value == "Automatic":
        df['Gearbox'].at[index] = int(3)
    elif value == "Semi-automatic":
        df['Gearbox'].at[index] = int(2)
```

```python
for index,value in df['Fuel type'].items():
    if "Electric" in value:
        df['Fuel type'].at[index] = "Electric"
    elif "Biogas" in value:
        df['Fuel type'].at[index] = "LPG"
    elif "Biodiesel" in value:
        df['Fuel type'].at[index] = "Diesel"
```

```python
# Trasformazione Fuel type in Nominal as follows
for index,value in df['Fuel type'].items():
    if value == "Electric":
        df['Fuel type'].at[index] = int(6)
    elif value == "Hybrid":
        df['Fuel type'].at[index] = int(5)
    elif value == "LPG":
        df['Fuel type'].at[index] = int(4)
    elif value == "Methane":
        df['Fuel type'].at[index] = int(3)
    elif value == "Diesel":
        df['Fuel type'].at[index] = int(2)
    elif value == "Gasoline":
        df['Fuel type'].at[index] = int(1)   
```

```python
# Trasformazione Emission class in Nominal as follows
for index,value in df['Emission class'].items():
    if value == "Euro 6":
        df['Emission class'].at[index] = int(6)
    elif value == "Euro 5":
        df['Emission class'].at[index] = int(5)
    elif value == "Euro 4":
        df['Emission class'].at[index] = int(4)
    elif value == "Euro 3":
        df['Emission class'].at[index] = int(3)
    elif value == "Euro 2":
        df['Emission class'].at[index] = int(2)
    elif value == "Euro 1":
        df['Emission class'].at[index] = int(1) 
```

```python
# Trasformazione Drivetrain in Nominal as follows Front = 1 & Rear = 0 & 4WD = 0.5
for index,value in df['Drivetrain'].items():
    if value == "Front":
        df['Drivetrain'].at[index] = int(0)
    elif value == "Rear":
        df['Drivetrain'].at[index] = int(0)
    elif value == "4WD":
        df['Drivetrain'].at[index] = int(1)
df.rename(columns={'Drivetrain': '4WD'}, inplace=True)
```

```python
# Trasformazione Upholstery type in Nominal as follows
for index,value in df['Upholstery'].items():
    if value == "Full leather" or value == "Other":
        df['Upholstery'].at[index] = int(5)
    elif value == "alcantara":
        df['Upholstery'].at[index] = int(4)
    elif value == "Part leather":
        df['Upholstery'].at[index] = int(3)
    elif value == "Cloth":
        df['Upholstery'].at[index] = int(2)
    elif value == "Velour":
        df['Upholstery'].at[index] = int(1) 
```

```python
for index, value in df['Country'].items():
    if value == " Altenerding" or value == " Schwab" or value == " Schweiz" or value == " Tschechien":
        df['Country'].at[index] = "DE"
```
